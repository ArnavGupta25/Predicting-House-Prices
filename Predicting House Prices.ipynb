## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Convert into DataFrame
df = pd.DataFrame(data)
#1 Calculation of Average Price:
average_price = df['price'].mean()
print(f"The average price of the real estate properties is: {average_price:.2f}")
````

````python
#2 Median price
print(df['price'].median())
````

````python
#3 Group by the "bedrooms" column and calculate the average price
average_price_by_bedrooms = df.groupby('bedrooms')['price'].mean().reset_index()
print(average_price_by_bedrooms.to_string(index=False))
````

````python
#4 Calculate the total price and count of properties for each "stories" category
custom_aggregations = {
    'price_total': ('price', 'sum'),
    'property_count': ('price', 'count')
}
result = df.groupby('stories').agg(**custom_aggregations)
print(result.to_string(index=False))
````

````python
#5 Create a custom category for high vs. low price
df['price_category'] = df['price'].apply(lambda x: 'High' if x > 10000000 else 'Low')
# Group by the new category and calculate average area
average_area_count_by_price_category = df.groupby('price_category').agg({'area': 'mean', 'price': 'count'}).reset_index()
average_area_count_by_price_category.rename(columns={'area': 'average_area', 'price': 'count'}, inplace=True)
print(average_area_count_by_price_category.to_string(index=False))
````

````python
#6 Group by both "stories" and "bedrooms" columns
custom_aggregations = {
    'average_price': ('price', 'mean'),
    'average_area': ('area', 'mean')
}
result = df.groupby(['stories', 'bedrooms']).agg(**custom_aggregations)
print(result)
````

````python
#7 Min and max price
print(df['price'].min(),df['price'].max())
````

````python
# 8. Count and print the number of houses with hot water heating.
hotwater_count = (df['hotwaterheating'] == 'yes').sum()
print(f"Number of houses with hot water heating: {hotwater_count}")
````

````python
# Handling Outliers & Outlier treatments

# Identify outliers using IQR
Q1 = data1['price'].quantile(0.25)
Q3 = data1['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

# Count the number of outliers
outliers = ((data1['price'] < lower_bound) | (data1['price'] > upper_bound)).sum()

print("Number of outliers:", outliers)
````

````python
# Replace outliers with 25th and 75th percentiles
data1['price'] = np.where((data1['price'] < lower_bound), Q1, data1['price'])
data1['price'] = np.where((data1['price'] > upper_bound), Q3, data1['price'])
````
